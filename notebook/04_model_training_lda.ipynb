```python
import pickle
from gensim.models import LdaModel
import warnings
warnings.filterwarnings("ignore")
```

```python
with open("../dataset/roblox_dictionary.pkl", "rb") as f:
    dictionary = pickle.load(f)

with open("../dataset/roblox_corpus_bow.pkl", "rb") as f:
    corpus_bow = pickle.load(f)

with open("../dataset/roblox_corpus_tfidf.pkl", "rb") as f:
    corpus_tfidf = pickle.load(f)
```

```python
NUM_TOPICS = 5
PASSES = 15
ITERATIONS = 400
RANDOM_STATE = 42
```

```python
lda_bow = LdaModel(
    corpus=corpus_bow,
    id2word=dictionary,
    num_topics=NUM_TOPICS,
    passes=PASSES,
    iterations=ITERATIONS,
    random_state=RANDOM_STATE,
    alpha='auto',
    eta='auto'
)
```

```python
lda_tfidf = LdaModel(
    corpus=corpus_tfidf,
    id2word=dictionary,
    num_topics=NUM_TOPICS,
    passes=PASSES,
    iterations=ITERATIONS,
    random_state=RANDOM_STATE,
    alpha='auto',
    eta='auto'
)
```

```python
for idx, topic in lda_bow.print_topics(num_words=10):
    print(f"Topik {idx+1}: {topic}")
```

```python
for idx, topic in lda_tfidf.print_topics(num_words=10):
    print(f"Topik {idx+1}: {topic}")
```

```python
lda_bow.save("../dataset/lda_bow_model")
lda_tfidf.save("../dataset/lda_tfidf_model")
```
