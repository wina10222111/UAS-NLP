```python
import pickle
from gensim.models import LdaModel
from gensim.models.coherencemodel import CoherenceModel
```

```python
# Load dictionary
with open("../dataset/roblox_dictionary.pkl", "rb") as f:
    dictionary = pickle.load(f)

# Load corpus
with open("../dataset/roblox_corpus.pkl", "rb") as f:
    corpus_bow = pickle.load(f)

with open("../dataset/roblox_corpus_tfidf.pkl", "rb") as f:
    corpus_tfidf = pickle.load(f)

# Load clean token hasil preprocessing
with open("../dataset/roblox_clean_text.pkl", "rb") as f:
    clean_text = pickle.load(f)
```

```python
lda_bow = LdaModel.load("../dataset/lda_bow_model")
lda_tfidf = LdaModel.load("../dataset/lda_tfidf_model")
```

```python
coherence_bow = CoherenceModel(
    model=lda_bow,
    texts=clean_text,
    dictionary=dictionary,
    coherence='c_v'
).get_coherence()

coherence_tfidf = CoherenceModel(
    model=lda_tfidf,
    texts=clean_text,
    dictionary=dictionary,
    coherence='c_v'
).get_coherence()

coherence_bow, coherence_tfidf
```

```python
lda_bow.log_perplexity(corpus_bow)
```

```python
lda_tfidf.log_perplexity(corpus_tfidf)
```
