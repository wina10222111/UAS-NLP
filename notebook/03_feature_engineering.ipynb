```python
import pickle
from gensim import corpora
from gensim.models import TfidfModel
```

```python
with open("../dataset/roblox_clean_text.pkl", "rb") as f:
    clean_text = pickle.load(f)
```

```python
clean_text[:3]
```

```python
clean_text = [
    [token for token in doc if len(token) > 2]
    for doc in clean_text
]
```

```python
dictionary = corpora.Dictionary(clean_text)

dictionary.filter_extremes(
    no_below=10,   # muncul minimal di 10 dokumen
    no_above=0.5  # maksimal di 50% dokumen
)
```

```python
corpus_bow = [dictionary.doc2bow(text) for text in clean_text]
```

```python
corpus_bow[:2]
```

```python
tfidf_model = TfidfModel(corpus_bow)
corpus_tfidf = tfidf_model[corpus_bow]
```

```python
with open("../dataset/roblox_dictionary.pkl", "wb") as f:
    pickle.dump(dictionary, f)

with open("../dataset/roblox_corpus_bow.pkl", "wb") as f:
    pickle.dump(corpus_bow, f)

with open("../dataset/roblox_corpus_tfidf.pkl", "wb") as f:
    pickle.dump(list(corpus_tfidf), f)
```
