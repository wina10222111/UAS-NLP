```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../dataset/ROBLOX_REVIEWS.csv")
```

```python
df = df[['review_text']]
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.shape
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def clean_text(text):
    text = str(text).lower()                  
    text = re.sub(r'[^a-z\s]', '', text)    
    tokens = text.split()                    
    tokens = [w for w in tokens if w not in stop_words]
    return tokens
```

```python
df['clean_review'] = df['review_text'].apply(clean_text)
df.head()
```

```python
df['clean_review'].map(len).describe()
```

```python
df = df[df['clean_review'].map(len) > 0]
df.shape
```

```python
df.to_csv("../dataset/clean_roblox_reviews.csv", index=False)
```
